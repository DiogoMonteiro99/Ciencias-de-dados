```python
import pandas as pd
import numpy as np
import plotly.express as px
import matplotlib.pyplot as plt
import seaborn as sn
df = pd.read_csv('Banks.csv')
df.head(10)
```

```python
df.shape
```

```python
def analise(dados):
    print(r'Tipo da coluna:',dados.dtypes)
    print(r'Quantidades de campos nulos:',dados.isna().sum())
    print(r'Quantidades de campos unicos: ',dados.describe())
        
```

```python
for coluna in df.columns:
   if df[coluna].dtypes == 'object':
      analise(df[coluna])
  
        
```

```python
for coluna in df.columns:
   if df[coluna].dtypes == 'int64' or df[coluna].dtypes == 'float64':
      analise(df[coluna])
```

```python
analise(df['text'])
```

```python
def grafico(dados1,tipo='',title=''):
    if tipo == 'box':
        fig = px.box(dados1,title=title)
        fig.show()
```

```python
for coluna1 in df.columns:
    if df[coluna1].dtypes == 'int64' or df[coluna1].dtypes == 'float64':  
        grafico(df[coluna1],'box','Analise do outliers')
        
               
```

```python
df['star'] = df['star'].fillna(df['star'].mean())
df['like'] = df['like'].fillna(df['like'].mean())
```

```python
analise(df['star'])
grafico(df['star'],'box','Analise de outliers')
```
